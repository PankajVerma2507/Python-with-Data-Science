```python
fig , ax = plt.subplots(nrows=2,ncols=2, figsize=(10,10), facecolor="red")
plt.show()
```

```python
# random line plot
plt.figure(figsize=(15,6),dpi =300)
x = np.linspace(0, 10, 100)
y = np.random.randn(100)
#print(x,y)
_ = plt.plot(x,y, linewidth =1, color='red', marker='o',)
```

```python
# random bar plot
plt.figure(figsize=(15,4),dpi=300)
x = np.random.randint(0, 10, 25)
y = np.random.randint(0, 10, 25)
plt.grid(False)
_ = plt.bar(x, y, color='red')
```
